```python
# define alpha and matrix for stable node
alpha = -0.25
matrix = {"title":"stable node", 
          "matrix": np.array([[alpha, 0],[0, -1]])} 

# show the plot
plot_phase_portrait(alpha, matrix)
```

```python
# define alpha and matrix for stable focus
alpha = -0.25
matrix = {"title":"stable focus", 
          "matrix": np.array([[alpha, alpha],[1/4, 0]])} 

# show the plot
plot_phase_portrait(alpha, matrix)
```

```python
# define alpha and matrix for unstable saddle
alpha = 0.25
matrix = {"title":"unstable saddle", 
          "matrix": np.array([[alpha, -alpha],[-1/4, 0]])} 

# show the plot
plot_phase_portrait(alpha, matrix)
```
